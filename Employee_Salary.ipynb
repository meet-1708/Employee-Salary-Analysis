```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('final.csv')
df.describe(include='object')
```

```python
# Explore the distribution of salaries using a histogram

plt.hist(df['salary'], bins=20)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

```python
# Explore the relationship between salary and experience level
correlation_exp_salary = df.groupby('experience_level')['salary'].mean()
print(correlation_exp_salary)

print()
# Investigate if there's a correlation between salary and employee residence
correlation_residence_salary = df.groupby('employee_residence')['salary'].mean()
print(correlation_residence_salary)

print()
# Check if there's a correlation between salary and job title
correlation_job_salary = df.groupby('job_title')['salary'].mean()
print(correlation_job_salary)
```

```python
# Compare average salaries across different job titles using a bar chart
import matplotlib.pyplot as plt

avg_salary_by_job_title = df.groupby('job_title')['salary'].mean().sort_values()
plt.figure(figsize=(10, 15))
plt.barh(avg_salary_by_job_title.index, avg_salary_by_job_title, color='skyblue')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title')
plt.show()
```

```python
# Visualize salary distributions based on experience levels using box plots
plt.figure(figsize=(10, 6))
experience_levels = df['experience_level'].unique()
boxplot_data = [df[df['experience_level'] == level]['salary'] for level in experience_levels]
plt.boxplot(boxplot_data, labels=experience_levels)
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.title('Salary Distribution by Experience Level')
plt.xticks(rotation=45)
plt.show()
```

```python
# Explore the relationship between salary and employee residence using a scatter plot
plt.figure(figsize=(10, 6))
employee_residences = df['employee_residence'].unique()
for residence in employee_residences:
    plt.scatter(df[df['employee_residence'] == residence]['employee_residence'], df[df['employee_residence'] == residence]['salary'], alpha=0.5, label=residence)
plt.xlabel('Employee Residence')
plt.ylabel('Salary')
plt.title('Salary vs. Employee Residence')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Plot bar charts to visualize the mode of categorical variables

# Job Title
plt.figure(figsize=(10, 6))
mode_job_title = df['job_title'].mode()[0]
job_title_counts = df['job_title'].value_counts()
job_title_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.title('Frequency of Job Titles')
plt.axhline(y=job_title_counts[mode_job_title], color='red', linestyle='--', label=f'Mode ({mode_job_title})')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Experience Level
plt.figure(figsize=(10, 6))
mode_experience_level = df['experience_level'].mode()[0]
experience_level_counts = df['experience_level'].value_counts()
experience_level_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Experience Level')
plt.ylabel('Frequency')
plt.title('Frequency of Experience Levels')
plt.axhline(y=experience_level_counts[mode_experience_level], color='red', linestyle='--', label=f'Mode ({mode_experience_level})')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Employment Type
plt.figure(figsize=(10, 6))
mode_employment_type = df['employment_type'].mode()[0]
employment_type_counts = df['employment_type'].value_counts()
employment_type_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Employment Type')
plt.ylabel('Frequency')
plt.title('Frequency of Employment Types')
plt.axhline(y=employment_type_counts[mode_employment_type], color='red', linestyle='--', label=f'Mode ({mode_employment_type})')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Work Setting
plt.figure(figsize=(10, 6))
mode_work_setting = df['work_setting'].mode()[0]
work_setting_counts = df['work_setting'].value_counts()
work_setting_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Work Setting')
plt.ylabel('Frequency')
plt.title('Frequency of Work Settings')
plt.axhline(y=work_setting_counts[mode_work_setting], color='red', linestyle='--', label=f'Mode ({mode_work_setting})')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Employment Type
mode_employment_type = df['employment_type'].mode()[0]
employment_type_counts = df['employment_type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(employment_type_counts, labels=employment_type_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Employment Type Distribution')
plt.legend(title='Mode', labels=[f'{mode_employment_type} (Mode)' if label == mode_employment_type else label for label in employment_type_counts.index])
plt.show()

# Work Setting
mode_work_setting = df['work_setting'].mode()[0]
work_setting_counts = df['work_setting'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(work_setting_counts, labels=work_setting_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'orange'])
plt.title('Work Setting Distribution')
plt.legend(title='Mode', labels=[f'{mode_work_setting} (Mode)' if label == mode_work_setting else label for label in work_setting_counts.index])
plt.show()
```
